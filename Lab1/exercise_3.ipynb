```python
import pandas as pd
import numpy as np

df = pd.read_csv('Exercise_3_Advanced_Regional_Sales_Data.txt', sep=',')

# Simulate sales data for multiple months by multiplying with random factors
np.random.seed(42)  # For reproducibility
random_factors = np.random.uniform(0.8, 1.2, size=len(df))  # Generate a 1D array of random factors

# Multiply the 'Sales' column by the random factors
df['Simulated_Sales'] = df['Sales'] * random_factors

# Display the resulting DataFrame
print(df.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# Plot för både 'Sales' och 'Simulated_Sales'
for region in df['Region'].unique():
    region_data = df[df['Region'] == region]
    plt.plot(region_data['Month'], region_data['Sales'], label=f'{region} - Original Sales', marker='o')
    plt.plot(region_data['Month'], region_data['Simulated_Sales'], label=f'{region} - Simulated Sales', linestyle='--', marker='x')

# Lägg till etiketter och titel
plt.title('Försäljningens utveckling över tid för varje region')
plt.xlabel('Månad')
plt.ylabel('Försäljning')
plt.xticks(rotation=45)
plt.legend()

# Visa diagrammet
plt.tight_layout()
plt.show()
```

```python
import pandas as pd 

df = pd.read_csv('Exercise_1_Advanced_Monthly_Temperature_Data.txt', sep=',')

print(df.head())
```

```python
df_mean = df['Temperature'].mean()
print('Mean temperature:', df_mean)

df_std = df['Temperature'].std()
print('Standard deviation of temperature:', df_std)
```

```python
# Beräkna gränserna för avvikelser
lower_bound = df_mean - 2 * df_std
upper_bound = df_mean + 2 * df_std

# Filtrera ut avvikelser
outliers = df[(df['Temperature'] < lower_bound) | (df['Temperature'] > upper_bound)]
print('Avvikelser:')
print(outliers)
```

```python
import matplotlib.pyplot as plt

# Scatterplot för alla data
plt.scatter(df['Day'], df['Temperature'], label='Data', color='blue')

# Scatterplot för avvikelser
plt.scatter(outliers['Day'], outliers['Temperature'], label='Outliers', color='red')

# Lägg till etiketter och titel
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Temperature Scatterplot with Outliers')
plt.legend()

# Visa plotten
plt.show()
```

# Avvikelser kan påverka analysen genom att:

	1.	Förvränga medelvärde och standardavvikelse, vilket ger en missvisande bild av datan.
	2.	Störa statistiska tester genom att öka variansen eller skapa snedfördelningar.
	3.	Minska modellens prestanda i maskininlärning genom att modellen överanpassar sig till extrema värden.
	4.	Förvränga visualiseringar, vilket gör mönster svårare att se.
	5.	Leva till felaktiga beslut om de inte beaktas korrekt.

För att hantera detta kan man exkludera avvikelser, använda robusta metoder eller transformera datan.

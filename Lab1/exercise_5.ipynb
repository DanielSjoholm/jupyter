```python
import pandas as pd

with open ('Exercise_5_Advanced_Text_Analysis.txt', 'r') as file:
    text = file.read()

print(text)
```

```python
words = text.lower().split()

word_counts = {}

for word in words:
    if word in word_counts:
        word_counts[word] += 1 
    else:
        word_counts[word] = 1

print(word_counts)
```

```python
import matplotlib.pyplot as plt

# Sortera ord efter frekvens och ta de 10 vanligaste
sorted_word_counts = sorted(word_counts.items(), key=lambda item: item[1], reverse=True)[:10]

# Dela upp orden och deras frekvenser i separata listor
words, counts = zip(*sorted_word_counts)

# Skapa stapeldiagram
plt.figure(figsize=(10, 6))
plt.bar(words, counts, color='skyblue')
plt.xlabel('Ord')
plt.ylabel('Frekvens')
plt.title('De 10 vanligaste orden')
plt.xticks(rotation=45)
plt.show()
```

```python
import re

# Dela upp texten i meningar
sentences = re.split(r'[.!?]', text)

# Ta bort eventuella tomma meningar
sentences = [sentence.strip() for sentence in sentences if sentence.strip()]

# Hitta den längsta meningen baserat på antalet ord
longest_sentence = max(sentences, key=lambda sentence: len(sentence.split()))

# Skriv ut den längsta meningen och antalet ord i den
print(f"Längsta meningen: '{longest_sentence}'")
print(f"Antal ord: {len(longest_sentence.split())}")
```

```python
# Beräkna totalt antal ord i texten
total_words = sum(len(sentence.split()) for sentence in sentences)

# Skapa ett diagram
plt.figure(figsize=(8, 5))
plt.bar(['Antal meningar', 'Totalt antal ord'], [len(sentences), total_words], color=['skyblue', 'lightgreen'])
plt.xlabel('Kategori')
plt.ylabel('Antal')
plt.title('Relation mellan antal meningar och totalt antal ord')
plt.show()
```

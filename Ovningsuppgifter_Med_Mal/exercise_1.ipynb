```python
import pandas as pd 

df = pd.read_csv('Exercise_1_Monthly_Temperature_Data.txt', sep=',')

print(df.head())
```

```python

max_temp = df['Temperature'].max()
min_temp = df['Temperature'].min()


max_temp_day = df.loc[df['Temperature'] == max_temp, 'Day'].iloc[0]
min_temp_day = df.loc[df['Temperature'] == min_temp, 'Day'].iloc[0]


print(f'Max temperature: {max_temp} och är på dag {max_temp_day}')
print(f'Min temperature: {min_temp} och är på dag {min_temp_day}')
```

```python
import matplotlib.pyplot as plt

df.plot(x='Day', y='Temperature')

plt.show()
```

```python
import pandas as pd 

df = pd.read_csv('Exercise_3_Regional_Sales_Data.txt', sep=',')

print(df.tail())
```

```python
Total_sales = df['Sales'].sum()

print('Total sales: ', Total_sales)
```

```python
BestSales = df['Sales'].max()

best_region = df.loc[df['Sales'] == BestSales, 'Region'].values[0]

print(f'The best sales were in {best_region} with {BestSales} sales')
```

```python
import matplotlib.pyplot as plt

df_sorted = df.sort_values(by='Sales', ascending=False)

df_sorted.plot(x='Region', y='Sales', kind='bar')

plt.show()
```
